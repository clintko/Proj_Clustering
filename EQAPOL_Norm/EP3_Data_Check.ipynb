## Filename 

Interetation of `101_F6901PRY_C01.npy`

- `101` is the lab number
- `F6901PRY` is the subject ID
- `C01` is treatment `C` replicate `01`

```python
markers = ['FSC-A', 'SSC-A', 'CD3', 'CD4', 'CD8', 'IFN+IL2']
```

```python
f = '101_F6901PRY_C01.npy'
xs = np.load(f)
```

```python
xs.shape
```

## Filter out points with gross scatter outliers

```python
xs = xs[(xs[:,0] >= 0) & (xs[:,0] <= 262144) & (xs[:,1] >= 0) & (xs[:,1] <= 263144), :]
xs.shape
```

### Plot

```python
dims = [(0,1), (3,4), (2,5)]
n = len(dims)
c = min(n, 4)
r = (c + 3)//4
pts = 10000

plt.figure(figsize=(c*4, r*4))
for k, dim in enumerate(dims):
    plt.subplot(r, c, k+1)
    x, y = xs[:pts, dim[0]], xs[:pts, dim[1]]
    plt.scatter(x, y, s=1)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(markers[dim[0]], fontsize=14)
    plt.ylabel(markers[dim[1]], fontsize=14)
```
